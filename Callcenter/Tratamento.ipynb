```python
#it may be necessary to install openpyxl to properly read the data
#!pip install openpyxl

#importing pandas
import pandas as pd

#reading data into dataframe
df = pd.read_excel('dados_atendimento.xlsx')
```

### Data treatment

```python
#small function to help the first analysis of our data

def show_info(data):
    print('DATASET SHAPE: ', data.shape, '\n')
    print('-'*50)
    print('FEATURE DATA TYPES:')
    print(data.info())
    print('\n', '-'*50)
    print('NUMBER OF UNIQUE VALUES PER FEATURE:', '\n')
    print(data.nunique())
    print('\n', '-'*50)
    print('NULL VALUES PER FEATURE')
    print(data.isnull().sum())
  
show_info(df)
```

```python
#Removing accents on the columns, change names to lower case
df.columns = map(str.lower, df.columns)
df.columns = [col.replace(" ", "_").replace("ó","o").replace("á","a").replace("/","_").replace("ê","e").replace("ç","c").replace("í","i").replace("ã","a").replace("__","_") for col in df.columns]

#a few more ETL steps, changing typos, lower casing info
for i in range(len(df)):
    if df['area'][i] == 'Cancelametos':
        df.area[i] = 'Cancelamento'
    elif df['area'][i] == 'RECLAMAÇÕES':
        df.area[i] = 'Reclamações'

#further column rename to avoid misteps
df.rename({'respondido_-_resolvido_(s_n)': 'resolvido_(s-n)'}, axis=1, inplace=True)

#Removing unnecessary information
df['resolvido_(s-n)'] = df['resolvido_(s-n)'].replace(regex=[r'Sim - '], value='')
```

```python
#checking if area data is properly changed
df.area.value_counts().sort_values()

df.head()
    
```

```python
#saving to xlsx
df.to_excel('dados_atendimento_tratados.xlsx')
```
